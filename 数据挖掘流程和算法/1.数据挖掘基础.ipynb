### 1.数据挖掘发展史
- 1989年从数据库中发现知识（KDD）
- 1995年第一届国际知识发现与数据挖掘大会
- 1997第三届国际学术大会    
数据挖掘是知识发现中的核心部分，它是指从数据集合中自动抽取隐藏在数据中的那些由价值的信息过程。

### 2.数据挖掘的基本任务
#### （1）分类与回归
分类是一种对离散型随机变量建模或预测的方法，如：客户流失预测；而回归是通过建模来研究来研究变量之间的相互关系的密切程度、结构状态以及进行模型预测的工具，如财政收入预测等。
#### （2）聚类
聚类分析是指在预先不知道类别标签的情况下，根据信息的相似度原则进行信息聚集的一种方法。目的是使得类别内数据得“差异性”尽可能小（即“同质性”尽可能大），类别间“差异性”尽可能大。
#### （3）关联规则
关联模式挖掘旨在从大量的数据当中发现特征之间或数据之间的相互依赖关系。这种存在于给定数据集中的频繁出现的关联模式，又称为关联规则。
#### （4）智能推荐
智能推荐用于联系用户和信息，并利用信息分析用户的兴趣偏好，为用户推荐感兴趣信息。
#### （5）时间序列
对不同的时间下取得的样本数据进行挖掘，用于分析样本数据之间的变化趋势，如：股指预测、天气预测等。

### 3.数据挖掘的通用流程

- 数据抽取
    - 抽取数据的标准：相关性、可靠性、有效性
    - 衡量数据质量的标准：资料完整无缺，各类指标项齐全；数据准确无误，反映的都是正常状态下的水平。
    - 抽取方式：随机抽样、等距抽样、分层抽样、按起始顺序抽样、分类抽样
